```python
# elbow method
# importing KMeans from sklearn
from sklearn.cluster import KMeans
```

```python
# calculate the variable within cluster sum of squares
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# lets see the values of the variable
wcss
```

```python
plt.plot(range(1,11), wcss, "-o")
```

```python
# define x and y axis
plt.plot(range(1,11), wcss, "-o")
plt.xlabel("No. of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
# k = 3, should be good
# now segment the data
kmeans = KMeans(n_clusters = 2)
clusters = kmeans.fit_predict(X)
```

```python
# we can check which point each cluster corresponds to
clusters
```

```python
# Finally, lets plot the results
# plot first cluster
plt.scatter(X[clusters==0,0], X[clusters==0,1],s=5,label='cluster 0')
plt.scatter(X[clusters==1,0], X[clusters==1,1],s=5, label='cluster 1')

plt.legend()
plt.show()
```
